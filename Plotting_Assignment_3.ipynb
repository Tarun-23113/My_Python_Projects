```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/titanic.csv")
df.head()
```

```python
# Plot a line plot of the average fare paid per passenger class (Pclass vs average Fare).

x = df.groupby('Pclass')['Fare'].mean()

plt.plot(x.index, x.values, marker='^', ls=':', c='b', label='Average Fare')
plt.title('Pclass vs average Fare')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.legend()
plt.show()
```

```python
# Create a scatter plot showing the relationship between Age and Fare. Use markers and no connecting lines.

x = df[['Age', 'Fare']]

plt.plot(x['Age'], x['Fare'], marker='s', ls='', c='black', label='Fare')
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

```python
# Plot a bar chart showing the count of passengers in each passenger class (Pclass).

x = df['Pclass'].value_counts()

plt.bar(x.index, x.values)
plt.title('count of passengers')
plt.xlabel('Passenger Class')
plt.ylabel('Total Passengers')
plt.show()
```

```python
# Plot a histogram of passengers ages.

plt.hist(data=df, x='Age', bins=5)
plt.show()
```

```python
# Adjust the number of bins and add transparency (alpha).

plt.hist(data=df, x='Age', bins=8, alpha=0.5)
plt.show()
```

```python
# Create a pie chart showing the distribution of passengers by gender (Sex).


plt.pie(df['Sex'].value_counts(), labels=['Male', 'Female'], autopct='%1.1f%%')
plt.show()
```

```python
# Create a figure with 2 rows and 2 columns of subplots:

# Histogram of Age

# Bar chart of Pclass

# Pie chart of Sex

# Scatter plot of Age vs Fare

# Add appropriate titles and axis labels.

fig, ax = plt.subplots(nrows=2, ncols=2, layout='constrained', figsize=(7, 7))

ax[0][0].hist(data=df, x='Age', bins=5)
ax[0][0].set_title('Histogram of Age')

x = df['Pclass'].value_counts()

ax[0][1].bar(x.index, x.values)
ax[0][1].set_title('Bar chart of Pclass')

ax[1][0].pie(df['Sex'].value_counts(), labels=['Male', 'Female'], autopct='%1.1f%%')
ax[1][0].set_title('Pie chart of Sex')

ax[1][1].scatter(df['Age'], df['Fare'])
ax[1][1].set_title('Scatter plot of Age vs Fare')

fig.show()
```

```python
# Apply sns.set_theme() with a style of your choice.

sns.set_theme(style='darkgrid', font_scale=0.8)

x = df.groupby('Pclass')['Fare'].mean()

plt.plot(x.index, x.values, marker='^', ls=':', c='b', label='Average Fare')
plt.title('Pclass vs average Fare')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.legend()
plt.show()
```

```python
# Use sns.lineplot() to plot average fare over passenger classes (Pclass vs average Fare).

x = df.groupby('Pclass')['Fare'].mean().reset_index()

sns.lineplot(data=x, x='Pclass', y='Fare')
plt.show()
```

```python
# Use sns.scatterplot() to visualize the relationship between Age and Fare, using Sex as hue.

sns.scatterplot(data=df, x='Age', y='Fare', hue='Sex', style='Sex')
plt.show()
```

```python
# Use sns.barplot() to plot the mean Fare for each Pclass.

sns.barplot(data=df, x=df['Pclass'], y=df['Fare'])
plt.show()
```

```python
# Use sns.countplot() to show the number of passengers who survived (Survived).

sns.countplot(data=df, x='Survived', hue='Sex', palette='pastel')
plt.show()
```

```python
# Plot the age distribution with sns.histplot(), including KDE.

sns.histplot(data=df, x='Age', bins=5, kde=True, stat='count')
plt.show()
```

```python
# Plot a KDE plot of fare distribution using sns.kdeplot(), with different colors for survivors and non-survivors.

sns.kdeplot(data=df, x='Fare', hue='Survived', fill=True, cut=True)
plt.show()
```
