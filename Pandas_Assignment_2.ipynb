```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/netflix_titles.csv")
```

```python
# Task 1: Show first and last 5 rows of the dataset

display(df.head(5))
display(df.tail(5))
```

```python
# Task 2: display column names, data types, and shape of the dataset

display(df.columns)
display(df.dtypes)
display(df.shape)
```

```python
# Task 3: Use describe() and info() to get summary stats

display(df.describe())

display(df.info())
```

```python
# Task 4: Find how many movies vs TV shows (value_counts on 'type')

df['type'].value_counts()
```

```python
# Task 5: Use loc and iloc to display:
#         - 10 random movies
#         - 5 entries with 'TV Show' type and 'Country' = 'India'

display(df[df['type'] == 'Movie'].sample(10))

display(df.query("type == 'TV Show' and country == 'India'").head(5))
```

```python
# Task 6: Check missing values in all columns (isnull().sum())

df.isnull().sum()
```

```python
# Task 7: Fill missing 'director' values with "Unknown"
#         Fill missing 'country' values with "Unspecified"

df['director'].fillna("Unknown")
df['country'].fillna("Unspecified")
```

```python
# Task 8: Drop rows where 'title' or 'type' is missing

df.dropna(subset=['title', 'type'], inplace=True)
```

```python
# Task 9: Remove duplicate rows (if any)

df.drop_duplicates(inplace=True)
```

```python
# Task 10: Extract year from 'date_added' and store in a new column 'year_added'
#          Hint: Convert 'date_added' to datetime using pd.to_datetime()

df['year_added'] = df['date_added'].str.extract(r'([^,]+)$')
```

```python
# Task 11: Rename 'listed_in' column to 'genre' for clarity

df.rename(columns={'listed_in':'genre'}, inplace=True)
```

```python
# Task 12: Sort dataset by 'release_year' descending and show top 10 oldest titles

df.sort_values('release_year', ascending=False)['title'].tail(10)
```

```python
# Task 13: Group by 'type' and 'release_year' → count how many titles per year

df.groupby(['release_year', 'type'])['title'].count()
```

```python
# Task 14: Create a new column 'is_in_India' → True if 'country' contains 'India'

df['is_in_India'] = df['country'] == 'India'
df
```

```python
# Task 15: Use pd.cut() to bin 'release_year' into 5 intervals (like decades)

df['Era'] = pd.cut(df['release_year'], bins=5, labels=['Beginning', 'Experimental', 'Golden', 'Saturated', 'Cinephiles'])
df[30:50]
```

```python
# Task 16: Use str.extract() to extract the first genre from the 'genre' column

df['genre'].str.extract(r'^([^,&]+)')
```

```python
# Task 17: Count top 5 directors with most titles

df.groupby('director')['title'].count().sort_values(ascending=False).head(5)
```

```python
# Task 18: Pivot Table → Show count of Movies/TV Shows by country (rows) and type (columns)

pd.pivot_table(df, 'title', 'country', 'type', aggfunc='count')
```

```python
# Task 19: Use groupby().transform() to create column:
#          'title_count_by_country' = number of titles per country

df['title_count_by_country'] = df.groupby('country')['title'].transform('count')
df
```

```python
# Task 20: Use melt() to reshape 'duration' and 'rating' into one column

pd.melt(df, id_vars='title', value_vars=['duration', 'rating'], var_name='D_R', value_name='Result')
```

```python
# Goal: Generate insights about Netflix content trends
```

```python
# Step 1: Clean 'duration' column (e.g. extract minutes or seasons)

df['duration_length'] = df['duration'].str.extract(r'^([^\s]+)')
df['duration_type'] = df['duration'].str.extract(r'^[0-9]+\s([A-z]+)')
df.head()
```

```python
# Step 2: Find top 10 countries with most Netflix content

df.groupby('country')['title'].count().sort_values(ascending=False).head(10)
```

```python
# Step 3: Analyze most common genres globally

df.groupby('genre')['title'].count().sort_values(ascending=False).head(15)
```

```python
# Step 5: Save final DataFrame to 'netflix_cleaned.csv'

df.to_csv("/content/netflix_cleaned.csv")
```
