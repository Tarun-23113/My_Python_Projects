```python
df['Investors'] = df['Investors'].str.strip()
```

```python
df['funding_total_usd'] = df['funding_total_usd']*1000000
```

```python
# 3.Renaming Cols for better readability
```

```python
df.rename(columns={
    'category_list':'sector',
    'funding_total_usd':'funding',
    'country_code':'country',
    'state_code':'state',
    'founded_at':'year',
    'Acquirer_Category':'Acquirer_Sector'
}, inplace=True)
```

```python
# 4.Checking if all dtypes are appropriate.
```

```python
mask = df['year'].str.isdigit()
df[~mask]
```

```python
df = df[mask].copy()
df
```

```python
df['funding_rounds'] = df['funding_rounds'].astype(int)
df['funding'] = df['funding'].astype(int)
df['Number_of_Investors'] = df['Number_of_Investors'].astype(int)
df['year'] = df['year'].astype(int)
```

```python
# 5.Creating New Features into Data
```

```python
# Making status as categorical

df['status'] = df['status'].str.strip()
df['status'] = df['status'].str.lower()
df['status'] = pd.Categorical(df['status'], categories=['ipo', 'acquired', 'operating', 'closed'], ordered=True)
```

```python
# First Step : what is most likely fortune of a startup

status_count = df['status'].value_counts()
plt.pie(status_count, labels=status_count.index, autopct="%1.2f%%")
plt.show()
```

```python
# Observing the trends and relation between the entities that directly leads to success
# plt.figure(figsize=(20, 8))
# plt.title('')
# plt.xlabel('')
# plt.ylabel('')
# plt.grid(ls='--', alpha=0.4)
# plt.legend()
# plt.show()
```

```python
# Line Plot of Sector wise success rate

x = pd.crosstab(index=df['sector'], columns=df['status'], normalize='index')*100

plt.figure(figsize=(20, 8))
plt.plot(x.index, x.values, marker='o', label=x.columns)
plt.title('Sector vs Status (rate of success)')
plt.xlabel('Sectors')
plt.ylabel('Rate')
plt.xticks(rotation=30)
plt.grid(ls='--', alpha=0.4)
plt.legend()
plt.show()
```

```python
# Mean of Fundings Sector wise

plt.figure(figsize=(20, 8))
sns.barplot(df.sort_values('sector'), x='sector', y='funding', hue='sector', palette='viridis')
plt.title('Average Funding per sector')
plt.xlabel('Sectors')
plt.ylabel('Avg. Funding')
plt.xticks(rotation=30)
plt.grid(ls='--', alpha=0.4)
plt.show()
```

```python
# Combined Relation of both above entities

x = pd.crosstab(index=df['sector'], columns=df['status'], values=df['funding'], aggfunc='median')
x.fillna(0, inplace=True)

plt.figure(figsize=(20, 8))
plt.plot(x.index, x.values, ls='--', lw=1, marker='o', label=x.columns)
plt.title('Sector vs Median Funding')
plt.xlabel('Sectors')
plt.ylabel('Median Funding')
plt.xticks(rotation=30)
plt.grid(ls='--', alpha=0.4)
plt.legend()
plt.show()
```

Above graphs gives us a hint that,


1.   Startups with specific sectors have shown a good success rate and less failure
2.   Second graph suspects the reason as funding to different startups could be reason and   
Yes, Some sectors are funded more than others

1.   Result we can see that in last plot, startups with higher fundings have listed in ipo or either acquired ( which both are considered a good end )






```python
# A distribution of status wise funding to re verify above claims

# Lets make a violin plot of funding vs status

# Funding is so skewed so we need to normalize it

df['log funding'] = np.log1p(df['funding'])

plt.figure(figsize=(20, 8))
sns.violinplot(df, x='status', y='log funding', hue='status', palette='viridis')
plt.title('Distribution of Fundings vs Status')
plt.xlabel('Startup\'s Status')
plt.ylabel('funding = e^y - 1')
plt.grid(ls='--', alpha=0.4)
plt.show()
```

Yes, we can say that those claims are true as the companies listed in IPOs and acquired have more funding than Operating Ones and Failed Ones

Check what leads to more funding

```python
# Relation between no of Investors vs Status

x = pd.crosstab(index=df['Number_of_Investors'], columns=df['status'], normalize='index')*100
x = x.head(15)

plt.figure(figsize=(20, 8))
plt.plot(x.index, x.values, ls='--', lw=1, marker='o', label=x.columns)
plt.title('No. of Investors vs Success Rate')
plt.xlabel('No. of Investors')
plt.ylabel('Rate')
plt.xticks(np.linspace(0, 15, 16))
plt.grid(ls='--', alpha=0.4)
plt.legend()
plt.show()
```

```python
# Relation between Funding Rounds vs Status

x = pd.crosstab(index=df['funding_rounds'], columns=df['status'], normalize='index')*100

plt.figure(figsize=(20, 8))
plt.plot(x.index, x.values, ls='--', lw=1, marker='o', label=x.columns)
plt.title('Funding Rounds vs Success Rate')
plt.xlabel('Funding Rounds')
plt.ylabel('Rate')
plt.xticks(np.linspace(0, 20, 21))
plt.grid(ls='--', alpha=0.4)
plt.legend()
plt.show()
```

Above graphs also help to claim that,



1.   Startups with more no. of investors had more success rate and less failing rate, till 15 no. of investors.
2.   Startups with more funding rounds had less chance to fail till 11 rounds, after that no startup was succeeded.


---

Hence, this can be seen as more the investors, more they helps to make startup successful but more than 15 can lead to mis management.

```python
# Year wise success distribution

plt.figure(figsize=(20, 8))
sns.kdeplot(df, x='year', hue='status', palette='viridis')
plt.title('Year wise distribution of Startup Status')
plt.xlabel('Year')
plt.ylim(0.00, 0.01)
plt.grid(ls='--', alpha=0.4)
plt.show()
```
