```python
import pandas as pd
df = pd.read_csv("/content/train.csv")
```

```python
# Task 2: Print column names and data types

df.dtypes
```

```python
# Task 3: Check shape, use describe() on numeric columns

df.shape

df.describe()
```

```python
# Task 4: Select rows where 'Age' > 60 using iloc and loc both

df.loc[:][df['Age'] > 60]
```

```python
# Task 5: Convert 'Pclass' to string using astype

df['Pclass'].astype(str)
df.dtypes
```

```python
# Task 6: Find how many missing values are in each column

df.isnull().sum()
```

```python
# Task 7: Fill missing 'Age' values with the mean

df['Age'].fillna(value=df['Age'].mean())
```

```python
# Task 8: Drop rows where 'Embarked' is missing

df['Embarked'].dropna()
```

```python
# Task 9: Remove duplicates in the dataset (if any)

df.drop_duplicates()
```

```python
# Task 10: Extract titles (Mr, Mrs, Miss) from 'Name' using regex with str.extract()

df['Titles'] = df['Name'].str.extract(r',\s*([A-Za-z]+)\.')
df
```

```python
# Task 11: Rename 'Sex' column to 'Gender'

df.rename(columns={'Sex':'Gender'}, inplace=True)
df
```

```python
# Task 12: Sort the dataset by 'Fare' in descending order

df.sort_values('Fare')
```

```python
# Task 13: Group by 'Pclass' and show average 'Age' and 'Fare'

df.groupby('Pclass')[['Age','Fare']].mean()
```

```python
# Task 14: Create a column 'Is_Child' where Age < 18 = True, else False

df['Is_Child'] = df['Age'] < 18
df.head(15)
```

```python
# Task 15: Create age groups using pd.cut() in 5 bins

label = ['Child', 'Teenager', 'Adult', 'Middle Age', 'Old Age']
df['Age_Group'] = pd.cut(df['Age'], bins=5, labels=label)
df
```

```python
# Task 16: Replace 'male' with 1 and 'female' with 0 in 'Sex' column using map()

mask = {'male': 1, 'female': 0}
df['Gender'] = df['Gender'].map(mask)
df
```

```python
# Task 17: Extract last names from 'Name' using .str.split() or regex

df['Last_Name'] = df['Name'].str.extract(r'^([^,]+)')
df
```

```python
# Task 18: Create a pivot table: average 'Fare' per 'Sex' and 'Pclass'

Pivot = pd.pivot_table(df, values='Fare', index='Gender', columns='Pclass', aggfunc='mean')
Pivot
```

```python
# Task 19: Normalize 'Fare' column group-wise by 'Pclass' using groupby().transform()
import numpy as np

df['Normalized'] = df.groupby('Pclass')['Fare'].transform(np.log1p)
df
```

```python
# Task 20: Use melt() to unpivot 'SibSp' and 'Parch' into a single column

pd.melt(df, id_vars=['PassengerId'], value_vars=['SibSp', 'Parch'], var_name='Mixed', value_name='count')
```

```python
df2 = pd.read_csv('/content/train.csv')
df2.head(20)
```

```python
# Step 1: Drop unnecessary columns ('PassengerId', 'Ticket', 'Cabin')

df2.drop(columns=['PassengerId', 'Ticket', 'Cabin'], inplace=True)
df2.head(20)
```

```python
# Step 2: Create 'Title' feature from 'Name'

df2['Title'] = df['Name'].str.extract(r',\s*([A-Za-z]+)\.')
df2
```

```python
new_order = list(df2.columns)
new_order.remove('Title')
new_order = ['Title'] + new_order
df2 = df2[new_order]
df2.head(20)
```

```python
# Step 3: Encode categorical features ('Sex', 'Embarked', 'Title')
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df2['Sex'] = le.fit_transform(df2['Sex'])
df2['Embarked'] = le.fit_transform(df2['Embarked'])
df2['Title'] = le.fit_transform(df2['Title'])
df2.head(20)
```

```python
# Step 4: Analyze survival rate by:
#    - Gender
#    - Class
#    - Title
#    - Age Group (use qcut or cut)

Sex = pd.crosstab(df2['Sex'], df2['Survived'], normalize=True)*100
display(Sex)
Class = pd.crosstab(df2['Pclass'], df2['Survived'], normalize=True)*100
display(Class)
Title = pd.crosstab(df2['Title'], df2['Survived'], normalize=True)*100
display(Title)
df2['Age_Group']  = pd.qcut(df2['Age'], q=3, labels=['Teenager', 'Adult', 'Old'])
Age_Group = pd.crosstab(df2['Age_Group'], df2['Survived'], normalize=True)*100
display(Age_Group)
```
