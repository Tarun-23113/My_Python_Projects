```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/company_employees.csv")
df.head()
```

```python
sns.countplot(df, x='joining_year', hue='gender', palette='viridis')
plt.xticks(rotation=30, fontsize=7)
plt.show()
```

```python
sns.countplot(df, x='gender', hue='gender', palette='magma')
plt.show()
```

```python
sns.barplot(df, x='department', y='performance_score', hue='department', palette='magma')
plt.show()
```

```python
sns.barplot(df, x='department', y='salary', hue='department', palette='magma')
plt.show()
```

```python
sns.lineplot(df, x='age', y='experience_years', errorbar='sd')
plt.show()
```

```python
sns.scatterplot(df, x='performance_score', y='salary')
plt.show()
```

```python
sns.histplot(df, x='salary', bins=10, kde=True, stat='count', hue='gender')
plt.show()
```

```python
sns.kdeplot(df, x='salary', fill=True, hue='gender', cut=True)
plt.annotate(text='Highest', xy=(60000, 1e-5), xytext=(62000, 1e-5), arrowprops=dict(arrowstyle='->'), fontsize=13)
plt.show()
```

```python
sns.boxplot(df, x='department', y='salary', notch=True)
plt.show()
```

```python
sns.violinplot(df, x='department', y='salary')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.xticks(rotation=30)
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.catplot(df, x='department', y='salary', kind='box')
```

```python
sns.pairplot(df, vars=['department', 'salary', 'joining_year'])
```
