```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/train.csv")
df
```

```python
# plt.plot() Versatility (Line and Scatter)

x = df['PassengerId'].dropna().head(100)
y = df['Fare'].dropna().head(100)

plt.figure(figsize=(12, 8))
plt.plot(x, y, color='purple', marker='^', linestyle=':', linewidth=1, label='Fare')
plt.title('Fare Trend for First 100 Passengers')
plt.xlabel('Passenger ID')
plt.ylabel('Fare')
plt.grid(linestyle='--', alpha=0.3)
plt.legend()
plt.show()
```

```python
# Create a scatter plot of Age (x) vs. Fare (y) for a random sample of 200 passengers

x = df[['Age', 'Fare']].dropna().sample(200)

plt.figure(figsize=(12, 7))
plt.plot(x['Age'], x['Fare'], marker='s', linestyle='', color='orange', linewidth=1, alpha=0.5, label='Age-Fare Data')
plt.title('Age vs Fare of Random 200 Samples(Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(linestyle=':', alpha=0.2)
plt.legend()
plt.show()
```

```python
# Easy Assignment: Bar Chart of Passenger Class Counts

x = df['Pclass'].value_counts().sort_index()

plt.figure(figsize=(5, 4))
plt.bar(x.index, x.values, label='Passenger Count', color='r')
plt.title('Passenger Counts by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passenger')
plt.grid(axis='y', linestyle=':', alpha=0.3)
plt.legend()
plt.show()
```

```python
# Distribution of Survival by Fare using Pie Chart and Histogram

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].pie(df['Survived'].value_counts(), labels=['Suvived', 'Not Survived'], autopct='%1.1f%%')
ax[0].set_title('Survival Proportions')

ax[1].hist(df['Fare'], bins=50, label='Fare')
ax[1].set_title('Fare Distribution')
ax[1].legend()

fig.show()
```

```python
# Two Subplots with Basic Labels and Grid

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11, 5))

ax[0].plot(df['PassengerId'].dropna().head(50), df['Fare'].dropna().head(50), ls=':', c='r', label='Fare over Passenger')
ax[0].set_title('Fare over Passenger ID')
ax[0].set_xlabel('Passenger ID')
ax[0].set_ylabel('Fare')
ax[0].grid(axis='y', alpha=0.3)
ax[0].legend()

x = df[['Age', 'Fare']].dropna().sample(100)
ax[1].scatter(x['Age'], x['Fare'], c='b', label='Age vs Fare')
ax[1].set_title('Age vs. Fare')
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Fare')
ax[1].legend()

fig.tight_layout()
fig.show()
```

```python
# Challenging Assignment: Subplots with Legend, Grid, and Saving

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 10), layout='constrained')

ax[0].hist(df['Age'], bins=20, label='Age Distribution', color='teal')
ax[0].legend()
ax[0].grid(axis='y', alpha=0.3)

ax[1].hist(df['Fare'], bins=30, label='Fare Distribution', color='salmon')
ax[1].legend()
ax[1].grid(axis='y', alpha=0.3)

fig.suptitle('Titanic Distributions')
fig.show()
fig.savefig("Titanic_Distributions.png", dpi=300)
```

```python
# Assignment 1: sns.histplot() (Distribution Plot)
# Goal: Understand the distribution of Age and Fare among passengers.

plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Passenger Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Fare Distribution by Survival Status

sns.histplot(data=df, x='Fare', bins=30, hue='Survived', stat='density')
plt.title('Fare Distribution by Survival Status')
plt.show()
```

```python
# Assignment 2: sns.countplot() (Categorical Plot)
# Goal: Visualize the counts of passengers based on their Sex and Pclass.

sns.countplot(data=df, x='Sex',  palette='pastel')
plt.title('Count of Passengers by Sex')
plt.show()
```
