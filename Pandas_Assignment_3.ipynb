```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("/content/matches.csv")
```

```python
# Task 1: View the first and last 5 rows, dataset shape and columns

display(df1.head())
display(df1.tail())
```

```python
# Task 2: Use info(), describe(), and value_counts() to inspect:
#         - Season-wise match count
#         - Most frequent venues
#         - Total matches per team

display(df1.info())
display(df1.describe())

display(df1['Season'].value_counts())
display(df1['venue'].value_counts())
display(pd.concat([df1['team1'], df1['team2']]).value_counts())
```

```python
# Task 3: List all unique teams and cities

display(df1['team1'].dropna(inplace=False).unique())
display(df1['city'].dropna(inplace=False).unique())
```

```python
# Task 4: Count how many times each team won the toss

display(df1['toss_winner'].value_counts().sort_values())
df1.groupby('toss_winner')['toss_winner'].count().sort_values()
```

```python
df2 = pd.read_csv("/content/deliveries.csv")
```

```python
# Task 5: Check for missing values in both datasets

display(df1.isnull())
display(df2.isnull())
```

```python
# Task 6: Drop rows with null values in 'winner' column (no result/abandoned)

df1['winner'].dropna(inplace=True)
```

```python
# Task 7: Clean column names by renaming 'player_of_match' → 'MOM'

df1.rename(columns={'player_of_match':'MOM', 'team1':'T1', 'team2':'T2', 'umpire1':'U1', 'umpire2':'U2', 'umpire3':'U3'}, inplace=True)
df1
```

```python
# Task 8: Create a new column 'win_by' in `matches`:
#         - "Runs" if 'win_by_runs' > 0
#         - "Wickets" if 'win_by_wickets' > 0
#         - "Tie/Other" otherwise

conditions = [df1['win_by_runs']>0,df1['win_by_wickets']>0]
choices = ["Runs", "Wickets"]
df1['win_by'] = np.select(conditions, choices, default="Tie/Other")
df1
```

```python
# Task 9: Which teams have the most wins overall?

df1['winner'].value_counts()
```

```python
# Task 10: Who won the most "Player of the Match" awards?

# df1.groupby('MOM').size().sort_values(ascending=False)
df1['MOM'].value_counts().sort_values(ascending=False)
```

```python
# Task 11: Group by season and find the most successful team per season

df1.groupby(['Season','winner']).size().reset_index(name='win_count').sort_values(['Season', 'win_count'], ascending=[True, False])
```

```python
# Task 12: Analyze the frequency of toss winners winning the match

t_m = df1[df1['toss_winner'] == df1['winner']]
t_m.groupby(['toss_winner', 'winner']).size()
```

```python
# Task 13: Create a pivot table with venue as index and count of wins per team

pd.pivot_table(df1, index='venue', columns='winner', aggfunc='count', fill_value=0)
# df1.groupby(['venue', 'winner']).size()
```
